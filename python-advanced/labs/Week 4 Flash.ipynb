```python
# Flash
# Write class Flash for modeling work with USB flash drive. Each Flash has limited capacity. The class should only allow to write files. The class constructor has a parameter capacity.

# Implement method write. First, it must check that the flash drive has enough memory to write this file, otherwise it need to throw the ValueError exception. If there is enough free memory, you need to write the file.
```

```python
class Flash:
    def __init__(self, capacity):
        """
        Constructor for Flash class
        
        Param: capacity - int|float, capacity of the flash drive inserted
        
        Returns: None
        """
        self.capacity = capacity

    def write(self, file_size):
        """
        Write a file to the flash drive

        Raises: ValueError if there is not enough space on the flash drive

        Returns: None
        """
        if file_size > self.capacity:
            raise ValueError("Damn, try a smaller file")
        self.capacity -= file_size
```
