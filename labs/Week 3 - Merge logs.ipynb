## Merge logs

You are given the results of the distributed message queue. Each consumer logged the received messages to its own file.  You need to merge the messages of all consumers into a single file, the messages should be sorted by date, if the dates match, sort additionally by consumer_id field.

<b>Input format:</b> The list of log files. Each log file is a list of JSONs with fields "date", "message" and "consumer_id"

<b>Output format:</b> Prepare messages in the desired order to the output list. Each line should contain fields "date" and "message" from original files separated by a tab (use character '\t').

Data examples:
- {"date": "2019-09-20 01:08:31", "message": "3ddx5=vkwxlz!fyu=.fh x5i6 onz%d q=#  &-", "consumer_id": 0}

```python
# You are given the results of the distributed message queue. Each consumer logged the received messages to its own file.  You need to merge the messages of all consumers into a single file, the messages should be sorted by date, if the dates match, sort additionally by consumer_id field.
# Input format: The list of log files. Each log file is a list of JSONs with fields "date", "message" and "consumer_id"

# Output format: Prepare messages in the desired order to the output list. Each line should contain fields "date" and "message" from original files separated by a tab (use character '\t').
import json
import pandas as pd

def solve(log_files : list) -> str:
    """
    Args:
        log_files: the list of log files

    Returns:
        list of formatted messages
    """
    # iterating over lines
    logs = list()
    for i in range(len(log_files)):
        logs += log_files[i]

    #parsing logs to list in a form of dicts
    sorted_logs = list()
    for i in logs:
           dict_log = json.loads(i)
           sorted_logs.append(dict_log)
    sorted_logs = sorted(sorted_logs, key=lambda d: (d['date'], d['consumer_id']))

    # 
    output = ['\t'.join([i['date'], i['message']]) for i in sorted_logs]

    # returning the string output
    return output
```
