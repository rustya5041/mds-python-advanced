## Complex numbers

Implement class <b>Complex</b> to work with complex numbers. To do this, implement methods:

1. class constructor and method str; Allowed types for `re` and `im` are `int and float`;
1. addition and subtraction operators with objects from classes `int, float and Complex`;
1. multiplication and division operators with objects from classes `int, float and Complex`;
1. comparison operator and method for calculating the modulus of a complex number.


- If you are not familiar with complex numbers, you can use <a href="https://en.wikipedia.org/wiki/Complex_number">this article</a> from Wikipedia
- If the argument type is not allowed raise TypeError
- The method `__init__` - save `re` and `im` to attributes, also check their types
- The method `__str__` should return a string in a special form. Examples: '7+4i', '0+0i', '12-89i', '0-2i'.
- Methods `__add__` and `__sub__` have to return new Complex numbers. Don't modify self object, methods `__iadd__` and `__isub__` are used for this purpose (they don't need to be implemented in this task)
- Methods `__add__` and `__sub__` have to return new Complex numbers too
- Method `__eq__` should return True if two Complex numbers are equal (equal real and imagine parts), False otherwise
- Method `__abs__` - the absolute value (or modulus or magnitude)
- What is '-> "Complex"' or '-> bool'? https://docs.python.org/3/library/typing.html

### Write the solution

```python
class Complex:
    # Part 1
    def __init__(self, re=0, im=0):
        """
        Class constructor. Takes the class params, checks for type
        
        re: int, float
        im: int, float

        raises: TypeError, if either argument is not either int or float
        """
        # check number of arguments passed to the class
        if not isinstance(re, (int, float)) or not isinstance(im, (int, float)):
            raise TypeError("Not int|float")
        self.re = re
        self.im = im
    
    def __str__(self) -> str:
        """
        Returns a string representation of the complex number

        returns: str
        """
        return f'{self.re}+{self.im}i' if self.im >= 0 else f'{self.re}-{abs(self.im)}i'
    
    # Part 2
    def __add__(self, other) -> "Complex":
        """
        Performs an addition of int, float, Complex

        other: int, float, Complex
        """
        if isinstance(other, (int, float)):
            return Complex(self.re + other, self.im)
        elif isinstance(other, Complex):
            return Complex(self.re + other.re, self.im + other.im)
        else:
            raise TypeError("Not int|float|Complex")
    
    def __sub__(self, other) -> "Complex":
        """
        This method is called when the - operator is used between two Complex objects. Substitutes numbers.

        other: int, float, Complex
        """
        if isinstance(other, (int, float)):
            return Complex(self.re - other, self.im)
        elif isinstance(other, Complex):
            return Complex(self.re - other.re, self.im - other.im)
        else:
            raise TypeError("Not int|float|Complex")

    # Part 3
    def __mul__(self, other) -> "Complex":
        """
        Performs a multiplication of int, float, Complex

        other: int, float, Complex
        """
        if isinstance(other, (int, float)):
            return Complex(self.re * other, self.im * other)
        elif isinstance(other, Complex):
            return Complex(self.re * other.re - self.im * other.im, self.re * other.im + self.im * other.re)
        else:
            raise TypeError("Not int|float|Complex")
    
    def __truediv__(self, other) -> "Complex":
        """
        Performs a division of int, float, Complex

        other: int, float, Complex
        """
        if isinstance(other, (int, float)):
            return Complex(self.re / other, self.im / other)
        elif isinstance(other, Complex):
            return Complex((self.re * other.re + self.im * other.im) / (other.re**2 + other.im**2), (self.im * other.re - self.re * other.im) / (other.re**2 + other.im**2))
        else:
            raise TypeError("Not int|float|Complex")

    # Part 4
    def __eq__(self, other) -> bool:
        """
        Checks if two Complex objects are equal

        other: Complex
        """
        if isinstance(other, Complex):
            return self.re == other.re and self.im == other.im
        else:
            raise TypeError("Not Complex")
    
    def __abs__(self) -> float:
        """
        Returns the absolute value of the complex number
        """
        return (self.re**2 + self.im**2)**0.5
```
