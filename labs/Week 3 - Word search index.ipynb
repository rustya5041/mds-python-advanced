## Word search index

Modern text assistants offer possible word completion options when you start typing a word. This process should be very fast, so you need to search the first letters of the word. For this purpose it is suggested to use one-letter and two-letter prefixes of words. Write a program that builds a hierarchical index from a list of words, where at the first level there is a prefix of length 1, at the second level there is a prefix of length 2, and at the next level there are full words (sort them before saving).

<b>Input format:</b> The list of words.

<b>Output format:</b> Return the resulting search index. Sort words at the last level in the index.

Input example:
- selection
- small
- none
- sales
- skin
- network
- sweep
- name
- stability
- news
- safety
- standard
- natural
- smell
- seriously

Output:

``{
    "s": {
        "sa": ["safety", "sales"],
        "sw": ["sweep"],
        "st": ["stability", "standard"],
        "sm": ["small", "smell"],
        "se": ["selection", "seriously"],
        "sk": ["skin"]
    },
    "n": {
        "ne": ["network", "news"],
        "na": ["name", "natural"],
        "no": ["none"]
    }
}
``

```python
def solve(words):
    """
    Returns the search_index based on the inputed word in a form of a 3-level dict, where first and second letters of the words are respective levels of the dict
    Args:
        words: list of words
    Returns:
        search_index
    """
        # creating a dictionary to store results
    index = dict()
    
    # adding the distinct letters to the dict
    for word in words:
        if word[0] not in index:  #frist letter
            index[word[0]] = dict()
        if len(word) > 1: 
            if word[0:2] not in index[word[0]]:
                index[word[0]][word[0:2]] = [] # slicing to get first to second let

        index[word[0]][word[0:2]].append(word)
        
    # sorting to stasify the condition
    for first_let in index:
        for second_let in index[first_let]:
            index[first_let][second_let].sort()
    
    # returning index
    return index
    
```
